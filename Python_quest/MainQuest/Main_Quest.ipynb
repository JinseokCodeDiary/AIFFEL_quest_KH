```python
# 문제 3-3. 클래스 업데이트
# 앞에서 구현했던 메서드들을 Kiosk 클래스에 추가합니다. 직접 코드를 작성해 보세요!

menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
price = [2000, 3000, 3000, 2500, 2500, 3000]

class Kiosk:
    def __init__(self):
        self.menu = menu
        self.price = price
        self.temp = ""
        self.menu_list = []
        self.price_list = []
        self.price_sum = 0

    # 메뉴 출력 메서드
    def menu_print(self):
        print(f"{'*' * 7} 메뉴 리스트 {'*' * 7}")
        for i in range (len(menu)):
            print(f"{i+1}. {menu[i] :<11}  --- {price[i]:>6} 원")
        return

    # 주문 메서드
    def menu_select(self):

        try:
            n = int(input("어떤 음료를 주문하시겠습니까? 숫자를 입력해 주세요."))
            while n > 0 and n <= 6:
                if 0 >= n or n > 6:
                    print("다시 입력해주세요.")
                    n = int(input("어떤 음료를 주문하시겠습니까? 숫자를 입력해 주세요."))
                t = input("Hot커피는 1번, Ice커피는 2번을 눌러주세요. ")
                while True:
                    if t == "1":
                        self.temp = "HOT"
                        break
                    elif t == "2":
                        self.temp = "ICE"
                        break
                    else:
                        print("다시 선택해주세요.")
                        t = input("Hot커피는 1번, Ice커피는 2번을 눌러주세요. ")

                self.menu_list.append(menu[n-1])
                self.price_list.append(price[n-1])
                self.price_sum += price[n-1]

                print(f"주문하신 내역은 {self.temp} {menu[n-1]}이고, 금액은 {price[n-1]} 입니다.")
                # print(self.menu_list, self.price_list, self.price_sum) 테스트 내용

                retry = input(f"주문음료: {len(self.menu_list)}개, 추가 주문 하시겠습니까? Y/N")
                while True:
                    retry = retry.upper()
                    if retry == "Y":
                        n = int(input("어떤 음료를 주문하시겠습니까? 숫자를 입력해 주세요."))
                        break
                    elif retry == "N":
                        n = 7
                        break
                    else:
                        print("다시 입력해 주세요.")
                        retry = input("추가 주문 하시겠습니까? Y/N")

        except ValueError as e:
            print("ValueError가 발생했습니다.", e)
        except TypeError as e:
            print("TypeError가 발생했습니다.", e)
        finally:
            print("수정 단계로 넘어갑니다.")

    def 주문수정(self):
        try:
            edit = input("어떻게 수정하시겠습니까? 1. 삭제, 2. 추가, 3. Nothing ")

            if edit == "1":
                delete = int(input("어떤 항목을 삭제하시겠습니까?"))
                print(self.menu_list[delete-1])
                self.menu_list.remove(self.menu_list[delete-1])
                self.price_sum = self.price_sum - self.price_list[delete-1]
                self.price_list.remove(self.price_list[delete-1])
                # print(self.menu_list, self.price_list, self.price_sum)
            elif edit == "2":
                print("어떤 항목을 추가하시겠습니까?")
                Kiosk.menu_select(self)
        except IndexError as e:
            print("인덱스 에러에요.", e)
        finally:
            print("지불 단계로 넘어갑니다.")


    # 지불
    def pay(self):
        payment = input("계산은 무엇으로 하시겠습니까? 1번 카드, 2번 현금")
        while payment != "1" or payment != "2":
            if payment == "1":
                print("1번, 카드를 선택하셨습니다. 카드를 삽입해주세요.")
                break
            elif payment == "2":
                print("2번, 현금을 선택하셨습니다. 종업원을 호출하겠습니다.")
                break
            else:
                print("다시 입력해주세요.")
                payment = input("계산은 무엇으로 하시겠습니까? 1번 카드, 2번 현금")

    # 주문서 출력
    def table(self):
        # 위
        print("-" * 32)

        # 중단
        print(f"ㅣ {' '*2} 계산서 {' '*16} ㅣ")
        for i in range(len(self.menu_list)):
            print(f"ㅣ {i+1}. {self.menu_list[i]:<12} --- {self.price_list[i]:>7}ㅣ")
        print(f"ㅣ {' '*2} 총액: {self.price_sum :>10} {' '*6} ㅣ")

        # 아래
        print("-" * 32)

        return
```

```python
a = Kiosk()         # 객체 생성
a.menu_print()      # 메뉴 출력
a.menu_select()     # 주문
a.table()           # 주문표 출력
a.주문수정()        # 주문 내역 수정
a.pay()             # 지불
a.table()           # 주문표 출력
```
