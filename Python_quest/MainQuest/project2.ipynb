```python
# 문제 3-3. 클래스 업데이트
# 앞에서 구현했던 메서드들을 Kiosk 클래스에 추가합니다. 직접 코드를 작성해 보세요!

# 클래스 정의
class Kiosk:
    def __init__(self):
        self.menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']     # 메뉴
        self.price = [2000, 3000, 3000, 2500, 2500, 3000]                                       # 가격
        self.menu_list = []             # 메뉴 리스트
        self.price_list = []            # 가격 리스트

    # 메뉴 출력 메서드
    def menu_print(self):
        print(f"{'*' * 7} 메뉴 리스트 {'*' * 7}")
        for i in range (len(self.menu)):
            print(f"{i+1}. {self.menu[i] :<11}  --- {self.price[i]:>6} 원")
        return

    # 주문 메서드
    def menu_select(self):

        try:
            n = int(input("어떤 음료를 주문하시겠습니까? 숫자를 입력해 주세요."))
            while n != 0:       # while문을 통해 0을 입력하면 빠져나가도록
                while n > len(self.menu) or n < 0:     # 메뉴 이외의 항목을 다시 입력하도록
                    print("다시 입력해주세요.")
                    n = int(input("어떤 음료를 주문하시겠습니까? 숫자를 입력해 주세요."))
                t = input("Hot커피는 1번, Ice커피는 2번을 눌러주세요. ")    # 온도 입력받기
                temp = ""

                while True:     # 1 입력하면 Hot, 2입력하면 Ice 꼭 1 or 2를 입력해야 빠져나갈 수 있다.
                    if t == "1":
                        temp = "HOT"
                        break
                    elif t == "2":
                        temp = "ICE"
                        break
                    else:
                        print("다시 선택해주세요.")
                        t = input("Hot커피는 1번, Ice커피는 2번을 눌러주세요. ")

                self.menu_list.append(self.menu[n-1])
                self.price_list.append(self.price[n-1])
                print(f"주문하신 내역은 {temp} {self.menu[n-1]}이고, 금액은 {self.price[n-1]} 입니다.")
                # print(self.menu_list, self.price_list) 테스트 내용

                retry = input(f"주문음료: {len(self.menu_list)}개, 추가 주문 하시겠습니까? Y/N")
                while True:     # while 추가 주문을 위한 반복
                    retry = retry.upper()   # 소문자로 입력해도 적용되도록
                    if retry == "Y":
                        n = int(input("어떤 음료를 주문하시겠습니까? 숫자를 입력해 주세요. ('0' 입력 - 다음 단계)"))
                        break       # break가 없을 경우 계속 반복됨
                    elif retry == "N":
                        n = 0       # 0을 입력하면 빠져나감.
                        break
                    else:
                        print("다시 입력해 주세요.")
                        retry = input("추가 주문 하시겠습니까? Y/N")

        except ValueError as e:         # ValueError 예외처리
            print("ValueError가 발생했습니다.", e)
        except TypeError as e:          # IndexError 예외처리
            print("TypeError가 발생했습니다.", e)
        finally:
            print("수정 단계로 넘어갑니다. \n")

    def 주문수정(self):
        try:
            edit = input("어떻게 수정하시겠습니까? 1. 삭제, 2. 추가, 3. Nothing ")
            if edit == "1":
                delete = int(input("어떤 항목을 삭제하시겠습니까? ('0' 을 입력하면 결제항목으로 이동합니다.)"))
                while delete != 0:
                    if self.menu_list[delete-1]:
                        print(f"주문 리스트에서 {self.menu_list[delete-1]}항목을 삭제합니다.")
                        self.menu_list.remove(self.menu_list[delete-1])
                        self.price_list.remove(self.price_list[delete-1])
                    self.table()
                    delete = int(input("어떤 항목을 삭제하시겠습니까? ('0' 을 입력하면 결제항목으로 이동합니다.)"))
                    # print(self.menu_list, self.price_list)
            elif edit == "2":
                print("어떤 항목을 추가하시겠습니까?")
                Kiosk.menu_select(self)
        except IndexError as e:
            print("인덱스 에러에요.", e)
        except ValueError as e:
            print("밸류 에러에요.", e)
        finally:
            print("지불 단계로 넘어갑니다. \n")

    # 지불
    def pay(self):
        payment = input("계산은 무엇으로 하시겠습니까? 1번 카드, 2번 현금")
        while payment != "1" or payment != "2":
            if payment == "1":
                print("1번, 카드를 선택하셨습니다. 카드를 삽입해주세요. \n")
                break
            elif payment == "2":
                print("2번, 현금을 선택하셨습니다. 종업원을 호출하겠습니다. \n")
                break
            else:
                print("다시 입력해주세요.")
                payment = input("계산은 무엇으로 하시겠습니까? 1번 카드, 2번 현금")

    # 주문서 출력
    def table(self):
        # 위
        print("-" * 32)

        # 중단
        print(f"ㅣ {' '*2} 계산서 {' '*16} ㅣ")
        for i in range(len(self.menu_list)):
            print(f"ㅣ {i+1}. {self.menu_list[i]:<12} --- {self.price_list[i]:>7}ㅣ")
        print(f"ㅣ {' '*2} 총액: {sum(self.price_list) :>10} {' '*6} ㅣ")

        # 아래
        print("-" * 32)

        return
```

```python
a = Kiosk()         # 객체 생성
a.menu_print()      # 메뉴 출력
a.menu_select()     # 주문
a.table()           # 주문표 출력
a.주문수정()        # 주문 내역 수정
a.pay()             # 지불
a.table()           # 주문표 출력
```
